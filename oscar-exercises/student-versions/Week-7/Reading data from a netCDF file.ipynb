```python
import matplotlib.pyplot as plt
import numpy as np

from scipy.io import netcdf
%matplotlib inline
```

```python
f = netcdf.netcdf_file('CMIP5/spatial_ncfiles/regrid_surf_tos_Oyr_IPSL-CM5A-LR_rcp85_r1i1p1.nc', 'r')
sst = f.variables['tos'][:]
lon = f.variables['longitude'][:]
lat = f.variables['latitude'][:]
time = f.variables['time'][:]
sst = np.ma.masked_greater(sst, 1000.)  # Missing values are encoded as 1.0e20; mask those.
```

```python
# This runs out to 2300 so we get a lot of warming...
fig, ax = plt.subplots(1, 1)
result = ax.pcolor(lon, lat, sst[-1, ...] - sst[0, ...], cmap='RdBu_r', vmin=-15., vmax=15.)
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')
plt.colorbar(result, label='Temperature Difference [K]')
```
